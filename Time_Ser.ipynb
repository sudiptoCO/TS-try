```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generating the Data

Here in this part we will be considering a timeseries data that follows the eq below

$$
y_t = 1.0001 \cdot y_{t-1} + \text{error}
$$


```python
date_index = pd.date_range(start='1/1/2022',end='31/12/2022',freq='D')
date_index
```

```python
# generating the values
values = [100]
for i in range(364):
    curr_value = 1.0001*values[-1] + np.random.normal(0,1,1)[0]
    values.append(curr_value)
```

```python
time_ser = pd.Series(values,date_index)
time_ser
```

```python
plt.figure(figsize=(10,4))
plt.plot(time_ser)
plt.ylabel("sales")
```

```python
detrended = []
for i in range(1,364):
    detrended.append(values[i]-values[i-1])
detrended_ser = pd.Series(detrended,date_index[1:-1])
```

```python
plt.figure(figsize=(10,4))
plt.plot(detrended_ser)
plt.ylabel("sales")
```
